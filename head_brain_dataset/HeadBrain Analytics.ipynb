```python
%matplotlib inline 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20.0,10.0)

data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)

n = len(x)

num = 0
den = 0
for i in range(n):
    num += (x[i] - mean_x) * (y[i] - mean_y)
    den += (x[i] - mean_x) ** 2
    
slope = num/den
constt = mean_y - (slope * mean_x)
print(slope, constt)
```

```python
max_x = np.max(x) + 100
min_x = np.min(x) - 100

x_r = np.linspace(min_x,max_x,1000)
y_r = constt + slope * x

plt.plot( x , y_r , color = '#58b970' , label = 'Regression Line')
plt.scatter( x , y , c = '#ef5423' , label = 'Scatter Plot')

plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(n):
    y_pre = constt + slope * x[i]
    ss_t += (y[i] - mean_y) ** 2
    ss_r += (y[i] - y_pre) ** 2

r2 = 1 - (ss_r/ss_t)
print(r2)
```
